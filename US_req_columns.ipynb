### Create the following new fields. 
1. Total Enrollment (TOT_ENR) = Enrollment including male and female students.
2. Create a new field called Average Enrollment (AVG_ENR) = Total Enrollment/7 (because there are 7 races).  
3. Total SATACT = total number of TOT_SATACT_M + TOT_SATACT_F
4. VAL_SAT = TOT_SATACT/TOT_ENR

````python
school_layout['TOT_ENR'] = school_layout['TOT_ENR_M'] + school_layout['TOT_ENR_F']
````

````python
school_layout['AVG_ENR'] = school_layout['TOT_ENR']/7
school_layout['TOT_SATACT'] = school_layout['TOT_SATACT_M'] + school_layout['TOT_SATACT_F']
school_layout['VAL_SAT'] = school_layout['TOT_SATACT']/school_layout['TOT_ENR']
````

````python
school_layout['VAL_WH'] = (school_layout['SCH_ENR_WH'] - school_layout['AVG_ENR']).abs()/school_layout['TOT_ENR']
school_layout['VAL_BL'] = (school_layout['SCH_ENR_BL'] - school_layout['AVG_ENR']).abs()/school_layout['TOT_ENR']
school_layout['VAL_HP'] = (school_layout['SCH_ENR_HP'] - school_layout['AVG_ENR']).abs()/school_layout['TOT_ENR']
school_layout['VAL_TR'] = (school_layout['SCH_ENR_TR'] - school_layout['AVG_ENR']).abs()/school_layout['TOT_ENR']
school_layout['VAL_HI'] = (school_layout['SCH_ENR_HI'] - school_layout['AVG_ENR']).abs()/school_layout['TOT_ENR']
school_layout['VAL_AS'] = (school_layout['SCH_ENR_AS'] - school_layout['AVG_ENR']).abs()/school_layout['TOT_ENR']
school_layout['VAL_AM'] = (school_layout['SCH_ENR_AM'] - school_layout['AVG_ENR']).abs()/school_layout['TOT_ENR']
````

### Calculate score based on the VAL where
1. 0.0-0.2 = 5 
2. 0.2-0.4 = 4
3. 0.4-0.6 = 3
4. 0.6-0.8 = 2
5. 0.8-1.0 = 1 

## TODO: 
1. change the scoring to be better reflective of the percentages

````python
school_layout['VAL_RACE'] = (school_layout['VAL_WH']+school_layout['VAL_BL']+\
    school_layout['VAL_HP']+school_layout['VAL_TR']+school_layout['VAL_HI']+\
    school_layout['VAL_AS']+school_layout['VAL_AM'])/7
````

````python
def score(school_layout):
    val = school_layout['VAL_RACE']
    if val>0 and val<=0.2:
        return 5
    elif val>0.2 and val<=0.4:
        return 4
    elif val>0.4 and val<=0.6:
        return 3
    elif val>0.6 and val<=0.8:
        return 2
    else:
        return 1

def score3(school_layout):
    val = school_layout['VAL_GENDER']
    if val>0 and val<=0.2:
        return 5
    elif val>0.2 and val<=0.4:
        return 4
    elif val>0.4 and val<=0.6:
        return 3
    elif val>0.6 and val<=0.8:
        return 2
    else:
        return 1
    
def score5(school_layout):
    val = school_layout['VAL_OFFENSES']
    if val>0 and val<=0.2:
        return 5
    elif val>0.2 and val<=0.4:
        return 4
    elif val>0.4 and val<=0.6:
        return 3
    elif val>0.6 and val<=0.8:
        return 2
    else:
        return 1
````

````python
def score2(school_layout):
    val = school_layout['VAL_AP_COURSES']
    if val>0 and val<=0.2:
        return 1
    elif val>0.2 and val<=0.4:
        return 2
    elif val>0.4 and val<=0.6:
        return 3
    elif val>0.6 and val<=0.8:
        return 4
    else:
        return 5

def score4(school_layout):
    val = school_layout['VAL_SATACT_COURSES']
    if val>0 and val<=0.2:
        return 1
    elif val>0.2 and val<=0.4:
        return 2
    elif val>0.4 and val<=0.6:
        return 3
    elif val>0.6 and val<=0.8:
        return 4
    else:
        return 5
````

````python
school_layout['RACE_SCORE'] = school_layout.apply(score, axis=1)
````

````python
school_layout['VAL_GENDER']=(school_layout['TOT_ENR_M'] - school_layout['TOT_ENR_F']).abs()/school_layout['TOT_ENR']
school_layout['GENDER_SCORE'] = school_layout.apply(score3, axis=1)
````

````python
school_layout['VAL_AP_COURSES'] = (school_layout['TOT_APENR_M'] + school_layout['TOT_APENR_F'])/school_layout['TOT_ENR']
school_layout['AP_SCORE'] = school_layout.apply(score2, axis=1)
school_layout['VAL_SATACT_COURSES'] = (school_layout['TOT_SATACT_M'] + school_layout['TOT_SATACT_F'])/school_layout['TOT_ENR']
school_layout['SATACT_SCORE'] = school_layout.apply(score4, axis=1)
school_layout['COURSES_SCORE'] = (school_layout['AP_SCORE']+school_layout['AP_SCORE'])//2
````

````python
school_layout['TOT_ARR'] = (school_layout['TOT_DISCWODIS_ARR_M'] + school_layout['TOT_DISCWODIS_ARR_F'])
school_layout['TOT_EXP'] = school_layout['TOT_DISCWODIS_EXPWE_M'] + school_layout['TOT_DISCWODIS_EXPWE_F'] +school_layout['TOT_DISCWODIS_EXPWOE_M'] + school_layout['TOT_DISCWODIS_EXPWOE_F']
school_layout['TOT_HARR'] = school_layout['TOT_HBDISCIPLINED_SEX_M'] + school_layout['TOT_HBDISCIPLINED_SEX_F'] + school_layout['TOT_HBDISCIPLINED_RAC_M'] + school_layout['TOT_HBDISCIPLINED_RAC_F']+ school_layout['TOT_HBDISCIPLINED_DIS_M'] + school_layout['TOT_HBDISCIPLINED_DIS_F']
school_layout['TOT_SUSP'] = school_layout['TOT_DISCWODIS_ISS_M'] + school_layout['TOT_DISCWODIS_ISS_F']
school_layout['VAL_OFFENSES'] = (school_layout['TOT_ARR']*0.4 + school_layout['TOT_EXP']*0.2 + school_layout['TOT_SUSP']*0.1  + school_layout['TOT_HARR']*0.3)
school_layout['OFFENSES_SCORE'] = school_layout.apply(score5, axis=1)
````

````python
school_layout['SCHOOL_SCORE'] = (school_layout['RACE_SCORE'] + school_layout['GENDER_SCORE'] + school_layout['COURSES_SCORE']
    + school_layout['OFFENSES_SCORE'])//4.0
````
